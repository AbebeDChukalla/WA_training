## 1. Does RS-derived Precipitation data agree with in-situ measurement?
### Example: Extract point time series from raster maps
Sometimes, we can obtain precipitation measurement from in-situ meteorological monitoring station or from global database (For example, NOAA Global Historical Climatology Network). This data can be used to validate remotely sensed precipitation data.
First, we'll read data from a text file or spread sheet. Using python code, this can be done systematically. For example, below is a code to read data from the given csv file. 

```python
import numpy as np
import os
import pandas as pd

ts_fh=r"E:\Exercise\2.3\Jordan\BEER SHEVA.csv"

# Read station name
Station_name=os.path.split(ts_fh)[-1].split(".")[0]
print(Station_name)

# Read metadata (coordinates)
f = open(ts_fh,'r') #open file to read
first=f.readline().split(";")
LatLon=[float(first[1]),float(first[3])]
f.close()
print(LatLon)

# Read data
obs_data=np.loadtxt(ts_fh,delimiter=';',skiprows=1,usecols=(1,2,4))
obs_ts=pd.DataFrame(obs_data,columns=['Year','Month','Data'])
obs_ts.head()
```

For a selected month and year, the observed value can be obtained from the dataframe by *.loc* method.

```python
year=2009
month=1
obs_ts.loc[obs_ts['Year']==year].loc[obs_ts['Month']==month]
```

When we know the coordinates of the station, we can extract the corresponding value in a raster map (RS data) for later comparison. Below is the example function *PixelCoordinates* to get corresponding row and column index of a point in raster mapset given the geographical coordinate.

```python
import glob
from osgeo import gdal
def PixelCoordinates(lon,lat,fh):
    """
    Find the corresponding pixel to a latitude and longitude.
    
    Parameters
    ----------
    lon : float or int
        Longitude to find.
    lat : float or int
        Latitude to find.
    fh : str
        Filehandle pointing to the file to be searched.
        
    Returns
    -------
    xpixel : int
        The index of the longitude.
    ypixel : int
        The index of the latitude.
    """
    SourceDS = gdal.Open(fh, gdal.GA_ReadOnly)
    xsize = SourceDS.RasterXSize
    ysize = SourceDS.RasterYSize
    GeoT = SourceDS.GetGeoTransform()
    assert (lon >= GeoT[0]) & (lon <= GeoT[0] + xsize * GeoT[1]), 'longitude is not on the map'
    assert (lat <= GeoT[3]) & (lat >= GeoT[3] + ysize * GeoT[5]), 'latitude is not on the map'
    location = GeoT[0]
    xpixel = -1
    while location <= lon:
        location += GeoT[1]
        xpixel += 1
    location = GeoT[3]
    ypixel = -1
    while location >= lat:
        location += GeoT[5]
        ypixel += 1
    return xpixel, ypixel   

P_fhs=glob.glob(r'E:\Exercise\2.3\Jordan\P\*.tif')
fh=P_fhs[1]
print(fh)
year=int(os.path.split(fh)[1].split("_")[-1][0:4])
month=int(os.path.split(fh)[1].split("_")[-1][4:6])
print(year,month)
xy=PixelCoordinates(LatLon[1],LatLon[0],fh)
print(xy)
```

Once obtain the row and column index in raster array from the point coordinates, we can extract the value at the point from the array. For example, below is the code to get the remotely sensed value at Beer Sheva station from the Precipitation raster file.

```python
def OpenRasterAsArray(in_fh,nan_value=True,bandnumber=1,dtype='float32'):  
    SourceDataSet = gdal.Open(in_fh, gdal.GA_ReadOnly)    
    Subdataset = SourceDataSet.GetRasterBand(bandnumber)
    NDV = Subdataset.GetNoDataValue()
    if NDV is None:
        NDV=-9999
    Array = Subdataset.ReadAsArray().astype(dtype)
    if nan_value:
        Array[Array == NDV] = np.nan
    return Array

map_array=OpenRasterAsArray(fh)
px_value=map_array[xy[1]][xy[0]]
print(px_value)
```

### Example: Compare RS-derived time series with observation
Sometimes, observation and remote sensing data have different available period. Thus, it's necessary to match the 2 timeseries by time. For example, below is how we match observed data with remote sensing data. 

```python
obs_values=[]
for i in range(len(rs_ts)):
    month=rs_ts.iloc[i]['Month']
    year=rs_ts.iloc[i]['Year']
    obs_value=-9999
    #check if the month and year has observed data
    for j in range(len(obs_ts)):
        jline=obs_ts.iloc[j]
        if (year == jline['Year']) & (month == jline['Month']):        
            obs_value=jline['Data']    
        else:
            continue 
    obs_values.append(obs_value)
rs_ts['Observed']=obs_values
rs_ts.head()    
```

When there's remote sensing data but no observed data available (-9999 value), these values should be dropped out of analysis.

```python
rs_ts.replace(-9999,np.nan)
rs_ts.drop(rs_ts.loc[rs_ts['Observed']==-9999].index,inplace=True)
rs_ts.head()
```

One common method to validate remotely sensed data with observed data is using scatter plot and regression line.

```python
from matplotlib import pyplot as plt
#plot scatter points
plt.scatter(rs_ts['Data'].values,rs_ts['Observed'].values)
plt.xlabel("WaPOR Precipitation [mm/month]")
plt.ylabel("Observed Precipitation [mm/month]")
plt.title(Station_name)

# regression line
#fitting RS and Observed data into a 1st order polynomial function (linear). z is the set of parameters of the fitted function
z = np.polyfit(rs_ts['Data'].values, rs_ts['Observed'].values, 1) 
#define the fitted linear function using parameter as p(x) where x is observed value
p = np.poly1d(z)
function = "%.2f * x+ (%.2f)"%(z[0],z[1]) 
#plot x=[0, max_x_value] and p(x)
plt.plot([0,max(rs_ts['Observed'])],p([0,max(rs_ts['Observed'])]),"r",label=function)
plt.legend(loc='upper left')

#adjust plot
plt.ylim(0,max(rs_ts['Observed']))
plt.xlim(0,max(rs_ts['Observed']))
plt.show()
```

## 2. Can RS-derived data close the water balance?
A csv of yearly P, ET, and dS derived from GRACE of Litani basin, Lebanon was prepared. Annual P, ET can be computed using the exampled code from previous exercises. The dS (Total storage change) is calculated from GRACE GFSC Mascon solution. For more details, you can have a look at the given scripts in *GRACE_for_basin* folder.
First, load the csv data as pandas Dataframe. The unit of the data values is MCM/year or Mm3/yr

```python
import pandas as pd
Litani_csv=r"E:\Exercise\2.3\Litani\P_ET_dS_stats.csv"
df=pd.read_csv(Litani_csv,sep=';',index_col=0)
df
```

We can calculate a new column as function of existing columns using the dataframe method *apply*

*df.apply(**lamda** row: function(row['column1'],row['column2'],...), axis=1)*

For example, below is how to calculate new column P-ET-dS


```python
df['P-ET-dS']=df.apply(lambda row: row['P'] - row['ET'] - row['dS'], axis=1)
df
```

Discharges were measured at the outlet of Litani basin and stored in a csv file. To compare the yearly discharge at outlet, we need to read the data from csv file and convert into yearly data.

```python
Qoutlet_csv=r"E:\Exercise\2.3\Litani\Qoutlet_litani.csv"
Qoutlet=pd.read_csv(Qoutlet_csv,sep=';',index_col=0,skiprows=1)
Qoutlet
```

Always check for nodata values and exclude them from calculation.

```python
Qoutlet.drop(Qoutlet.loc[Qoutlet['m3/s']==-9999].index,inplace=True)
Qoutlet
```

Using the *apply* method, convert discharge in m3/s to Mm3.

```python
Qoutlet['Mm3']=Qoutlet.apply(lambda row: (row['m3/s']*86400*row['days'])/(1000000), axis=1)
Qoutlet
```

Remember how to sum data by year?

```python
Qoutlet_y=Qoutlet.groupby(['year']).sum()
Qoutlet_y
```

Now we can compare Qoutlet with P-ET-dS of Litani basin. What does it mean when P-ET-dS is not close to Qoutlet?

```python
df['Qoutlet']=Qoutlet_y['Mm3'].values
from matplotlib import pyplot as plt
ax = plt.subplot(111)
ax.plot(df['P-ET-dS'].values,label='P-ET-dS')
ax.plot(df['Qoutlet'].values,label='Qoutlet')
ax.legend()
ax.set_xticklabels(df['year'].values)
ax.set_xlabel('Time')
ax.set_ylabel('Flows [MCM/year]')
plt.show()
```
